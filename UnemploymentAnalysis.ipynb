```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Unemployment in India.csv', encoding='latin1')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.dropna(subset=['date', 'estimated_unemployment_rate_'], inplace=True)
print(df.info())
print(df.head())
sns.set(style='whitegrid')
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='date', y='estimated_unemployment_rate_', hue='area', estimator='mean')
plt.title('Unemployment Rate Over Time by Area')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Area')
plt.tight_layout()
plt.show()
region_avg = df.groupby('region')['estimated_unemployment_rate_'].mean().sort_values()
plt.figure(figsize=(12, 18))
sns.barplot(x=region_avg.values, y=region_avg.index, palette='viridis')
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Average Rate (%)')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='date', y='estimated_labour_participation_rate_', hue='area', estimator='mean')
plt.title('Labour Participation Rate Over Time by Area')
plt.xlabel('Date')
plt.ylabel('Participation Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Area')
plt.tight_layout()
plt.show()
```
